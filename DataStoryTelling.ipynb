```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
def barplot(title, data, custom = 1, swap=False):
        fig, ax = plt.subplots()
        if swap == False:
            ax = sns.barplot(x=data.index, y=data.values)
            ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
        else:
            ax = sns.barplot(y=data.index, x=data.values)
            ax.set_yticklabels(ax.get_yticklabels())
        patches = ax.patches
        if swap == False:
            if custom == 1:
                for i in range(len(patches)):
                    x = patches[i].get_x() + patches[i].get_width()/2
                    y = patches[i].get_height()+0.1
                    label = "{:.1f}".format(y-0.1)
                    ax.annotate(label, (x,y), ha='center')
            else:
                for i in range(len(patches)):
                    x = patches[i].get_x() + patches[i].get_width()/2
                    y = patches[i].get_height()+custom
                    label = "{:.1f}".format(y-custom)
                    ax.annotate(label, (x,y), ha='center')
        else:
            if custom == 1:
                for i in range(len(patches)):
                    y = patches[i].get_y() + patches[i].get_height()/2
                    x = patches[i].get_width()+0.1
                    label = "{:.1f}".format(x-0.1)
                    ax.annotate(label, (x,y), ha='center')
            else:
                for i in range(len(patches)):
                    y = patches[i].get_y() + patches[i].get_height()/2
                    x = patches[i].get_width()+custom
                    label = "{:.1f}".format(x-custom)
                    ax.annotate(label, (x,y), ha='center')
        
        plt.title(title, loc='left')
        name = title.replace(' ', '_') + '.jpg'
        plt.savefig(name, bbox_inches="tight")
        plt.show()
        
        
def top_x_avg(df, by, what, top):
    average = df.groupby(by)[what].mean().sort_values(ascending=False)
    return average[:top]

def compare(first, second):
    how_many = []
    for i in range(len(first)):
        for j in range(len(second)):
            if first.index[i] == second.index[j]:
                how_many.append(first.index[i])
    count = len(how_many)
    return how_many, count
```
